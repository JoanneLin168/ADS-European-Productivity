```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
productivity = pd.read_csv("../Datasets/cleaned_Productivity.csv")
technology = pd.read_csv("../Datasets/TAI2000-2018_Overall.csv")
technology.head()

```

```python
productivity.head()
```

```python
# don't need the names and reformat
technology.drop(columns = "Country_Name", inplace=True)
technology.set_index("Country_Code", inplace=True)
technology
```

```python
#extract only the countries in both
technology = technology[technology.index.isin(productivity["LOCATION"])]
technology.index
```

```python
# write to csv
technology.to_csv('../Datasets/cleaned_technology.csv')
```

```python
for year in technology.columns:
    plt.scatter(technology[year], productivity[year], label = year)
plt.xlabel("technological advancement level")
plt.ylabel("productivity")
```

```python
# doesnt look too correlated

# work out average correlation between tech and productivity within years
productivity.drop(labels=["2019","2020","2021","mean"], axis=1, inplace=True)
productivity.set_index("LOCATION", inplace=True)

productivity.head()

```

```python
corrs = []
count = 0

for code in technology.index:
    corr = np.corrcoef(technology.loc[code],productivity.loc[code])[0][1]
    corrs.append(corr)
    count+=1

print(sum(corrs)/count)
```
