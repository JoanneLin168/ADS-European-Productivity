# EDA for the EU Economic Sentiment Indicator Dataset

```python
import numpy as np
from matplotlib  import pyplot as plt
import pandas as pd

df = pd.read_csv("../Datasets/cleaned_small_ESI.csv")
print(df)
```

```python
print(df.loc[2,:])
print(df["2022/04"])
```

```python
means = []
sums = []
vars = []
c = 0
for line in df:
    if c != 0:
        means.append(df[line].mean())
        sums.append(df[line].sum())
        vars.append(df[line].var())
    c += 1
cols = df.axes[1][1:len(df.axes[1])]
plt.plot(cols, means)
plt.xticks(rotation="vertical")
plt.xlabel("Month")
plt.ylabel("ESI")
plt.title("Average ESI Across Countries")
plt.show()
```

```python
plt.plot(cols,sums)
plt.xticks(rotation="vertical")
plt.xlabel("Month")
plt.ylabel("ESI")
plt.title("Sum of ESI Across Countries")
plt.show()
```

```python
plt.plot(cols,vars)
plt.xticks(rotation="vertical")
plt.xlabel("Month")
plt.ylabel("ESI")
plt.title("Variance of ESI Across Countries")
plt.show()
```

```python
for i in range(0,32):
    plt.plot(df.loc[i][1:,])
plt.legend(df["Economic Sentiment Per Month EU Countries"], bbox_to_anchor=(1,1))
plt.xticks(rotation="vertical")
plt.xlabel("Month")
plt.ylabel("ESI")
plt.title("ESI Across Countries")
plt.show()
```

```python
n = [7,8,11,17,19,28]
s = [4,9,10,16,22,25,31]
e = [0,3,5,13,14,20,21,24,26,27,29,30]
w = [1,2,6,12,15,18,23]
north = []
south = []
east = []
west = []
varN = []
varS = []
varE = []
varW = []
c = 0
for line in df:
    if c != 0:
        north.append(df[line][n].mean())
        varN.append(df[line][n].var())
        south.append(df[line][s].mean())
        varS.append(df[line][s].var())
        east.append(df[line][e].mean())
        varE.append(df[line][e].var())
        west.append(df[line][w].mean())
        varW.append(df[line][w].var())
    c += 1
plt.plot(cols, north)
plt.plot(cols, south)
plt.plot(cols, east)
plt.plot(cols, west)
plt.xticks(rotation="vertical")
plt.legend(["North", "South", "Central and East", "West"])
plt.xlabel("Month")
plt.ylabel("ESI")
plt.title("Mean ESI Across Geographical EU Regions (from EuroVoc)")
plt.show()
```

```python
plt.plot(cols, varN)
plt.plot(cols, varS)
plt.plot(cols, varE)
plt.plot(cols, varW)
plt.xticks(rotation="vertical")
plt.legend(["North", "South", "Central and East", "West"])
plt.xlabel("Month")
plt.ylabel("ESI")
plt.title("Variance of ESI Across Geographical EU Regions (from EuroVoc)")
plt.show()
```
