```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt



```

```python
entire_education = pd.read_csv("../Datasets/education_original.csv")
LFS = pd.read_csv("../Datasets/cleaned_LFS.csv")
productivity = pd.read_csv("../Datasets/cleaned_Productivity.csv")

entire_education.head()
```

```python
# drop columns that are constant in this data

data = entire_education.copy()

education = entire_education.drop(columns=["DATAFLOW", "LAST UPDATE", "freq", "sex", "age", "unit", "isced11"])
education.head()
```

```python
# the geographical labels are not the same
print(education["geo"].unique())
print(LFS["COUNTRY"].unique())
print(productivity["LOCATION"].unique())
```

```python
# drop aggregates and countries not in productivity
indicesToDrop = education[((education.geo == 'EU27_2020') | 
                       (education.geo == "ME") |
                       (education.geo == "RS") |
                       (education.geo == 'EA20'))].index
education.drop(indicesToDrop, inplace=True)
education.reset_index(inplace=True, drop=True)
print(education["geo"].unique())
```

```python
# agonisingly and painstakingly look through the dataset and relabel (turns out I didn't need to do this)

renaming = {"AT" : "AUT", "BE" : "BEL", "BG" : "BGR", "CH" : "CHE", "CY" : "CYP", "CZ" : "CZE",
            "DE" : "DEU", "DK" : "DNK", "EE" : "EST", "EL" : "GRC", "ES" : "ESP", "FI" : "FIN", 
            "FR" : "FRA", "HR" : "HRV", "HU" : "HUN", "IE" : "IRL", "IS" : "ISL", "IT" : "ITA",
            "LT" : "LTU", "LU" : "LUX", "LV" : "LVA", "MK" : "MKD", "MT" : "MLT", "NL" : "NLD",
            "NO" : "NOR", "PL" : "POL", "PT" : "PRT", "RO" : "ROU", "SE" : "SWE", "SI" : "SVN", 
            "SK" : "SVK", "TR" : "TUR", "UK" : "GBR"}

education.replace(renaming, inplace=True)

education
```

```python
# configure to same format as productivity
education = pd.pivot_table(education, index="geo", columns="TIME_PERIOD", values="OBS_VALUE")
education.head()
```

```python
# saves education to datasets folder
education.to_csv('../Datasets/cleaned_education.csv', index=False)
```

```python
# using nuno's code to see what it looks like (thanks nuno)

numberOfRows = education.shape[0]
numberOfColumns = education.shape[1]
plt.figure(figsize=(10, 7))

colors = plt.cm.hsv(np.linspace(0, 1, numberOfRows))

for row in range(numberOfRows):
    countryCode = education.iloc[row].name
    plt.plot(np.array(education.columns[0:]), education.iloc[row][0:], color=colors[row], label=countryCode)

plt.grid()
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='best')
plt.xlabel('Year') 
plt.ylabel('Percentage with at least upper secondary educational attainment')
plt.title("Percentage of the population with at least upper secondary educational attainment, age group 25-64")
```

```python
#drop columns not in education data
alignedProductivity = productivity.drop(["mean", "2000", "2001","2002","2003","2004"], axis=1)
alignedProductivity.set_index("LOCATION", inplace=True)
```

```python
# Choose the UK and plot them alongside each other
plt.figure(figsize=(10, 7))
plt.plot(np.array(education.columns[0:]), education.loc['GBR'][0:], label="education") # can change the code to
plt.plot(np.array(education.columns[0:]), alignedProductivity.loc['GBR'], label="productivity") # country if you want
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='best')
```

```python
corrs = []
count = 0

for code in education.index:
    try:
        corr = np.corrcoef(education.loc[code],alignedProductivity.loc[code])[0][1]
        if not (np.isnan(corr)):
            corrs.append(np.corrcoef(education.loc[code],alignedProductivity.loc[code])[0][1])
            count+=1
    except KeyError:
        # yeah guys I know its a try except but its 10:45pm and its the easiest way to
        # discount countries that are in the education dataset that arent in productivity
        print("A country was missed out")

print(sum(corrs)/count)
```
