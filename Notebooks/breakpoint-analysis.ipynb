# Breakpoint Analysis

This notebook aims to test out breakpoint detection and modelling methods

## Loading Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ruptures as rpt
```

```python
path = '../Data/'
data = np.load(f'{path}voltages.npy')[6000:11000]

plt.plot(data)
```

## Ruptures

```python
model = rpt.Window()
model.fit(data)
breaks = model.predict(n_bkps=2)
```

```python
breaks_rpt = []
for i in breaks:
    breaks_rpt.append(i)

plt.plot(data, label='data')
plt.title('Audience')
print_legend = True
for i in breaks_rpt:
    if print_legend:
        plt.axvline(i, color='red',linestyle='dashed', label='breaks')
        print_legend = False
    else:
        plt.axvline(i, color='red',linestyle='dashed')
plt.grid()
plt.legend()
plt.show()
```
