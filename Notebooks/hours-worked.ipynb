## Get predictor Dataset

Hours Actually Worked: https://stats.oecd.org/Index.aspx?DataSetCode=ANHRS

```python
dataPath = "../Data/hours-actually-worked.csv"
rawX = pd.read_csv(dataPath, header=0, nrows=None)
rawX
```

## Refactor Predictors to match Labour Productivity

```python
# Select between Total Employment or Dependant Employment
X = rawX.loc[rawX['EMPSTAT'] == "TE"]

# Remove all unneeded columns
dropLabels = ["EMPSTAT", "Country", "Employment status", "FREQUENCY", "Frequency", "Time", "Unit Code", "Unit", "PowerCode Code",
              "PowerCode", "Flag Codes", "Flags", "Reference Period Code", "Reference Period"]
for label in dropLabels:
    X = X.drop(label, axis=1)
X
```

```python
columnHeaders = ["LOCATION", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]
Xnew = pd.DataFrame(columns=columnHeaders)

for i, country in enumerate(X.COUNTRY.unique()):
    countryData = X.loc[X['COUNTRY'] == country]
    newRow = [country]
    newRow.extend(countryData.Value)
    while len(newRow) < 13:
        newRow.append(np.mean(countryData.Value))
    print(country)
    Xnew.loc[i] = newRow

Xnew
```
